# Data Structures and Algorithm - Classes

##### when you create a class you are creating your own data type.​(we're going to use classes for every one of our data​ structures.​‌)


```python
#After creating a class, you want Python to do particular things so you will use "Constructors"(it has a particular syntax) and "Destructors".
#Constructor example:
class Cookie:
    def __init__(self, color):  #This is the constructor method
        #If it has self in there, it is a method that is part of a class.​‌
        self.color = color  #be different about a method than a function is the self keyword
        
    def get_color(self): #if all it has is self that means there are no arguments passed to this
        return self.color
        #It's just going to return the color for the particular cookie that you're calling this on.​‌

    def set_color(self, color): #this is a setter method
        self.color = color  #this is going to set the color for the particular cookie object that you're calling this on.​‌

cookie_one = Cookie("Green")  #Creating an object of the Person class. cookie_one是一个Cookie类的对象,这个饼干是绿色的
cookie_two = Cookie("Blue")  #cookie_two是一个Cookie类的对象,这个饼干是蓝色的

print('Color of cookie_one:', cookie_one.get_color())  #Calling the get_color method on the cookie_one object
print('Color of cookie_two:', cookie_two.get_color())  #Calling the get_color method on the cookie_two object

cookie_one.set_color("Red")  #Changing the color of cookie_one to Red using the set_color method

print('\ncookie_one is now:', cookie_one.get_color())  #Calling the get_color
print('cookie_two is still:', cookie_two.get_color())  #cookie_two的颜色没有变
```

# Data Structures and Algorithm - Pointer

###### 两个数字变量指向同一数字内存，不能改变数字的值
###### 两个字典变量指向同一字典内存，可以改变字典的值
###### 如果两个变量指向同一个节点，就可以改变该节点的值

```python
#integers are what are called immutable, which means they cannot be​‌ changed.​
#Once you put a number 11 into a particular place in memory, you cannot change that number 11 
num1 = 11 #creating an integer 11 somewhere in memory, And then we point the variable num1 at that integer.​‌
num2 = num1

print("Before num2 value is updated:")
print("num1:", num1)  #11
print("num2:", num2)  #11

## id(num1) -> This is going to give us the address where num1 and num2 are stored.
print("\nnum1 points to:", id(num1))  #id()函数返回对象的内存地址
print("num2 points to:", id(num2))
# num1 and num2 are pointing to the same number 11 in memory.​‌

num1 = 22  #Now we are creating a new integer 22 somewhere else in memory, and we are pointing num2 at that new integer.​‌
print("After num1 value is updated:")
print("num1:", num1)  #22
print("num2:", num2)  #11

print("\nnum1 points to:", id(num1))  
print("num2 points to:", id(num2))



num2 = 33  #Now we are creating a new integer 22 somewhere else in memory, and we are pointing num2 at that new integer.​‌
print("After num2 value is updated:")
print("num1:", num1)  #22
print("num2:", num2)  #33

print("\nnum1 points to:", id(num1))  
print("num2 points to:", id(num2))
```

```python
dict1 = {'value': 11 }
dict2 = dict1  #dict2 is now pointing to the same dictionary as dict1
# dict1 and dict2 are both going to point to the same dictionary.

print("\nBefore dict2 value is updated:")
print("dict1=", dict1)
print("dict2=", dict2)

print("\ndict1 points to:", id(dict1))  
print("dict2 points to:", id(dict2))
#dict1 and dict2 are both pointing to the same address and memory

dict2['value'] = 22
print("\nAfter dict2 value is updated:")
print("dict1=", dict1)
print("dict2=", dict2)

print("\ndict1 points to:", id(dict1))  
print("dict2 points to:", id(dict2))


#dict3 = {'value': 33 }
#dict2 = dict3  #Now we are creating a new dictionary somewhere else in memory, and we are pointing dict2 at that new dictionary.​‌
#dict1 = dict2
# now you don't have a variable that's pointing to {'value':22}, you don't have any way to access this.
#So what Python does in this situation is Python will remove this through a process called garbage collection.​‌
```
